```python
type_map = {
     #"man": "meski",
     #"men": "meski",
     #"woman": "damski",
     #"women": "damski",
     #"boy": "chlopiecy",
     #"girl": "dziewczecy",
     #"uni": "unisex",
     #"agro": "rolnicze",
     "house": "dom",
     "room": "pokoj",
     "flat": "mieszkanie",
     #"fulltime": "pelny etat",
    #"efficient": "sprawny",
    "damaged": "uszkodzony",
     #"halftime": "pol etatu",
     #"parttime": "niepelny etat",
     "practice": "praktyki"
}

roomsize_map = {
    "one": "jednoosobowy",
    "two": "dwuosobowy",
    "three": "trzyosobowy"
}

rims_map = {
    "alu": "alufelgi"
}

contract_map = {
    "zlecenie": "zlecenie",
    "part": "czasowe",
    "selfemployment": "samozatrudnienie",
    "contract": "kontrakt"
}

requirements_map = {
    "student": "student",
    "disability": "niepełnosprawność"
}

petrol_map = {
    "petrol": "benzyna",
    "diesel": "diesel",
    "lpg": "lpg",
    "cng": "cng",
    "electric": "elektryczny"
}

car_body_map = {
    "sedan": "sedan",
    "estate-car": "kombi",
    "hatchback": "hatchback",
    "suv": "suv",
    "mvp": "mvp",
    "minibus": "bus",
    "cabriolet": "kabriolet",
    "off-road-vehicle": "terenowy",
    "coupe": "coupe",
    "pickup": "pickup"
}

furniture_map = {
    "yes": "meble",
    #"no": "stan surowy"
}

rooms_map = {
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4"
}

floor_map = {
    "parter": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4"
}

def params_converter(text):
    result = []
    pars = text.split('<br>')
    for par in pars:
        par2 = par.split('<=>')
        if par2[0] == 'size':
            result.append(par2[1].replace('s_', ''))
        elif par2[0] == 'type' and par2[1] in type_map:
            if par2[1] in type_map:
                result.append(type_map[par2[1]])
            else:
                result.append(par2[1].replace('-', ' '))
        #elif par2[0] == 'roomsize' and par2[1] in roomsize_map:
        #    result.append(roomsize_map[par2[1]])
        elif par2[0] == 'rims' and par2[1] in rims_map:
            result.append(rims_map[par2[1]])
        #elif par2[0] == 'contract' and par2[1] in contract_map:
        #    result.append(contract_map[par2[1]])
        elif par2[0] == 'requirements' and par2[1] in requirements_map:
            result.append(requirements_map[par2[1]])
        elif (par2[0] == 'width' or par2[0] == 'profile') and par2[1] != 'inny':
            result.append(par2[1].replace('w_', ''))
        elif par2[0] == 'drivinglicence' and par2[1] != 'other':
            result.append(par2[1])
        elif par2[0] == 'laptop_manufacturer' and par2[1] != 'pozostale':
            result.append(par2[1].replace('-', ' '))
        elif par2[0] == 'floor_select':
            result.append(par2[1].replace('plus', '').replace('floor_', 'pietro '))
        elif par2[0] == 'model' and par2[1] not in ['others', 'pozostale', 'inny']:
            result.append(par2[1].replace('-', ' '))
        #elif par2[0] == 'petrol' and par2[1] in petrol_map:
        #    result.append(petrol_map[par2[1]])
        #elif par2[0] == 'car_body' and par2[1] in car_body_map:
        #    result.append(car_body_map[par2[1]])
        #elif par2[0] == 'color' and par2[1] != 'inny':
        #    result.append(par2[1].replace('_', ' '))
        elif par2[0] == 'transmission' and par2[1] == 'automatic':
            result.append('automat')
        elif par2[0] == 'furniture' and par2[1] in furniture_map:
            result.append(furniture_map[par2[1]])
        elif par2[0] == 'rooms' and par2[1] in rooms_map:
            result.append(rooms_map[par2[1]])
        elif par2[0] == 'floor' and par2[1] in floor_map:
            result.append(floor_map[par2[1]])
        elif par2[0] == 'mark' and par2[1] != 'inny':
            result.append(par2[1].replace('-', ' '))
        elif par2[0] == 'axes' and par2[1] == '2':
            result.append("dwuosiowy")
        #elif par2[0] in ['tiretype', 'builttype']:
        #    result.append(par2[1])

    return ' '.join(result)

import re

stopwords = ['do','z','na', 'i', 'dla', 'w', 'now(a|y|e)', 'okazj(a|e)', 'sprzedam', 'jak', 'stan', 
            'od', 'gratis', 'gwarancj(a|i|e)', 'super', 'bardzo', 'nie', 'lub', 'oraz' , 
            'dobr(a|y|e)', 'wlasciciel', 'tani(o|y|e)*', 'wysylka', 'bez', 'cena', 'za', 'firmy', 'oryginal[a-z]*', 'promoc[a-z]+',
            'bdb', 'zamiana', 'polecam', 'price', 'new', 'duz(y|a|o|e)', 'po', 'rok', 'ze', 'r|roz|rozm|rozmiar', 
            'pod', 'lat', 'transport', 'piekn(y|a|e)', 'idealn(y|a|e)', '[0-9]+ *zl', '[0-9]+ *(szt|sztuki)',
             'stan', 'pl', 'polska']

slowo = {
    'vw': 'volkswagen',
    'karnister': 'kanister',
    'r': 'rozmiar',
    'roz': 'rozmiar',
    'rozm': 'rozmiar'
}

polskie = {
    'ą': 'a',
    'ę': 'e',
    'ć': 'c',
    'ł': 'l',
    'ń': 'n',
    'ó': 'o',
    'ź': 'z',
    'ż': 'z',
    'ś': 's'
}

def clean_text(dataset, field):
    dataset[field] = dataset[field].str.lower()
    dataset[field] = dataset[field].str.replace(r'<[a-z]*>', ' ')
    dataset[field] = dataset[field].str.replace('h&m', 'him')
    
    pl_repl = lambda m: polskie[m.group('litera')]
    dataset[field] = dataset[field].str.replace(r"(?P<litera>{})".format('|'.join(list(polskie.keys()))), pl_repl)
    
    #Remove non-ascii
    dataset[field] = dataset[field].str.replace(r"[^a-z0-9 ]", ' ')
    #dataset[field] = dataset[field].str.replace(r'( |^)([0-9]+)( |$)', ' ') #remove digits
    
    #bledy i skroty
    word_repl = lambda m: slowo[m.group('slowo')]
    
    dataset[field] = dataset[field].str.replace(r"((?<= )|^)(?P<slowo>{})((?= )|$)".format('|'.join(list(slowo.keys()))), word_repl)
#   for word in numbers:
#         dataset[field] = dataset[field].str.replace(r'( |^){}( |$)'.format(word[0]), word[1])
    dataset[field] = dataset[field].str.replace(r'((?<= )|^)({})((?= )|$)'.format('|'.join(stopwords)), ' ')
    dataset[field] = dataset[field].str.replace(r' +', ' ')

def preprocess_set(dataset, train = False):
    dataset['params'] = dataset['params'].apply(lambda row: params_converter(row))
    clean_text(dataset, 'title')

def save_set(dataset, path):
    dataset[['title', 'params']].to_csv(path, header=None, index=None, sep=' ', escapechar=' ', quoting=csv.QUOTE_NONE, mode='w')
```
